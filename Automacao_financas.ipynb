# Passo 1:

- Instalando o yfinance e plotando os dados das ações da empresa desejada

```python
!pip install yfinance
```

```python
import yfinance
```

```python
ticker = input("Digite a sigla da ação desejada: ")
dados = yfinance.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

# Passo 2

Análises dos últimos 6 meses:

- Cotação máxima
- Cotação minima
- Cotação atual


```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

- clicar no botão 'Nova mensagem'
- Preencher o email do destinatário

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("outlook.live.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")
pyautogui.click(x=218, y=171) ## Posição em que o botão 'nova mensagem' está posicionado

pyperclip.copy("gestor@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

mensagem = f"""
Prezado gestor,

seguem as análises dos ultimos seis meses da ação {ticker}:

Cotação máxima: R$ {round(maxima)}
Cotação mínima: R$ {round(minima)}
Cotação atual: R$ {round(atual)}
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=464, y=172) ## Posição em que o botão 'enviar' está posicionado
```
